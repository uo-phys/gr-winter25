```python
import numpy as np
import matplotlib.pyplot as plt

import sympy as sp
```

$$
t = v - r - 2M\log\left|\frac{r}{2M} - 1\right|
$$

Schwarzschild metric:
$$
ds^2 = -\left(1 - \frac{2M}{r}\right)dv^2 + 2dvdr + r^2(d\theta^2 + \sin^2\theta d\phi^2)
$$

Radial light cones:
ingoing:
$$
v = \mathrm{const.}
$$
outgoing:
$$
v - 2\left(r + 2M\log\left|\frac{r}{2M} - 1\right|\right) = \mathrm{const.}
$$

Define one more coordinate:
$$
\tilde{t} = v - r
$$

```python
def outgoing_light_t_tilde_over_M(r_over_M, r0=None):
    t_tilde_over_M = 0 if r0 is None else -(outgoing_light_t_tilde_over_M(r0))
    t_tilde_over_M += r_over_M + 4*np.log(np.abs(.5 * r_over_M - 1))
    return t_tilde_over_M
```

```python
rmax = 4
ngrid = 10

for r0 in np.linspace(2, rmax, ngrid):
    r_over_M = np.linspace(r0, rmax, 500)
    plt.plot(r_over_M, outgoing_light_t_tilde_over_M(r_over_M, r0), ls='--', color='k')

for r0 in np.linspace(2, 0, ngrid):
    r_over_M = np.linspace(0, r0, 500)
    plt.plot(r_over_M, outgoing_light_t_tilde_over_M(r_over_M, r0), ls='--', color='k')

for r0 in np.linspace(0, 2*rmax, ngrid):
    r_over_M = np.linspace(0, r0, 500)
    plt.plot(r_over_M, -r_over_M + r0, ls='--', color='k')

plt.xlabel(r'$r/M$')
plt.axvline(2, color='k')
plt.xlim(0, rmax)
plt.ylim(0, rmax)
plt.ylabel(r'$\tilde{t}/M$');
```
