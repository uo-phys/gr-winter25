# Schwarzschild Effective Potential

Recall from lecture that we found constants of motion associated with our Killing vectors:

$$
e = -\bm{\xi} \cdot \bm{u} = \left(1 - \frac{2M}{r}\right)\frac{dt}{d\tau} \\
\ell = \bm{\eta} \cdot \bm{u} = r^2 \sin^2\theta \frac{d\phi}{d\tau}
$$

and energy integral reminiscent of Newtonian mechanics, with

$$
\epsilon = \frac{e^2 - 1}{2} = \frac{1}{2}\left(\frac{dr}{d\tau}\right)^2 + V_\text{eff}(r)
$$

where the effective potential is

$$
V_\text{eff}(r) = -\frac{M}{r} + \frac{\ell^2}{2r^2} - \frac{M\ell^2}{r^3}.
$$

```python
def V_eff(r_over_M, ℓ_over_M=4.3, newtonian=False):
    """Function for computing the Schwarzschild effective potential for a given dimensionless radius and angular momentum per unit rest mass.  `newtonian=True` returns just the Newtonian terms."""
    Veff = -1/r_over_M + .5*ℓ_over_M**2/r_over_M**2
    if not newtonian:
        Veff += -ℓ_over_M**2/r_over_M**3
    return Veff
```

```python
r_over_M = np.linspace(2, 40, 500)

plt.plot(r_over_M, V_eff(r_over_M))
plt.plot(r_over_M, V_eff(r_over_M, newtonian=True), ls='--')
plt.ylim(-.05, .05)
plt.ylabel(r'$V_\mathrm{eff}$');
plt.xlabel(r'$\frac{r}{M}$');
```
